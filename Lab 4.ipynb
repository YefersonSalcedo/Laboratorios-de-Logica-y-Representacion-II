## Problema 1 (20 %)

Considerando un árbol binario, diseñe un algoritmo recursivo para imprimir todas las trayectorias desde cada nodo hoja hasta el nodo raíz. Una vez diseñado el algoritmo, diseñe un segundo algoritmo que realice el anterior proceso por fuerza bruta. ¿Cuáles son los costos computacionales temporales de ambos algoritmos?

```python
class Nodo:
  def __init__(self, valor):
    self.valor = valor
    self.izq = None
    self.der = None
```

```python
def camino_recursivo(raiz):
  if raiz is None:
    return

  trayectorias(raiz, [])

def trayectorias(raiz, trayectoria_actual):
  if raiz is None:
    return

  trayectoria_actual.append(raiz.valor)

  if raiz.izq is None and raiz.der is None:
    print(trayectoria_actual[::-1])
  else:
    if raiz.izq is not None:
      trayectorias(raiz.izq, trayectoria_actual.copy())
    if raiz.der is not None:
      trayectorias(raiz.der, trayectoria_actual.copy())
```

```python
def camino_fuerza_bruta(raiz):
  if raiz is None:
    return None

  trayectorias = []
  pila = [(raiz, [raiz.valor])]

  while pila:
    nodo, trayectoria_actual = pila.pop()

    if nodo.izq is None and nodo.der is None:
      trayectorias.append(trayectoria_actual[::-1])
    else:
      if nodo.der is not None:
        pila.append((nodo.der, trayectoria_actual + [nodo.der.valor]))
      if nodo.izq is not None:
        pila.append((nodo.izq, trayectoria_actual + [nodo.izq.valor]))

  return trayectorias
```

**Explicar el procedimiento que realiza los dos algoritmos:**

**Camino recursivo:** En este algoritmo se usan dos funciones, en la primera función se verifica que la raiz no sea nula(de ser nula no se hace nada y se imprime un valor nulo) para luego llamar a la segunda función llamada trayectorias; la función trayectorias usa dos parametros: la raiz y una lista vacia en la cual se almacenaran los caminos que se encuentren. como primer paso en la función trayectorias se agrega el valor de la raiz a la lista, luego se hace una validación para verificar si estamos en las hojas de arbol e imprimir la lista que tenemos en orden inverso, en caso de no estar en las hojas del arbol se inician a hacer 2 llamadas recursivas de la función, pasandole como argumento el nodo siguiente a la izquierda y el nodo siguiente a la derecha respectivamente, además, como segundo argumento le enviaremos una copia de la lista que tenemos actualmente.

**Camino fuerza bruta:** En este algoritmo inciamos haciendo la validación para verificar que la raiz no sea nula, luego creamos una lista vacia llamada trayectorias y una pila que contiene una tupla (la raiz y una lista para los caminos que encontremos), luego mediante un ciclo while que se ejecutara hasta que la pila este vacia, realizaremos la busqueda de las trayectorias hasta las hojas, dentro del bucle estraemos de la pila el ultimo nodo y la trayectoria para llegar a el, luego se realiza una validación para verificar si estamos en las hojas del arbol y en caso de estar en una hoja agregamos la trayectoria en orden inverso a la lista creada en un inicio llamada trayectorias; en caso de no estar en una hoja se agregará a la pila el hijo o los hijos del nodo en que estamos actualmente junto con la trayectoria para llegar hasta ellos, para luego continuar con el blucle; por último retornamos la lista llamada trayectorias.

```python
# Prueba

raiz = Nodo(1)
raiz.izq = Nodo(2)
raiz.der = Nodo(3)
raiz.izq.izq = Nodo(4)
raiz.izq.der = Nodo(5)
raiz.der.izq = Nodo(6)
raiz.der.der = Nodo(7)
raiz.der.izq.izq = Nodo(8)
raiz.der.izq.der = Nodo(9)

print(camino_recursivo(raiz))
print()
print(camino_fuerza_bruta(raiz))
```

## Problema 2 (15 %)

Dado un conjunto de LSLs previamente organizadas de menor a mayor cada una de ellas, diseñe un algoritmo basado en estructura de datos *heap* para unir las LSLs, considerando que cada una de ellas tiene diferentes longitudes.

**Pista 1:**

Pueden utilizar la propiedad *MinHeap* o *MaxHeap*

**Pista 2:**

$L1 = [10, 20, 30, 40]$

$L2 = [15, 25, 35]$

$L3 = [27, 29, 37, 48, 93]$

$L4 = [32, 33]$

$Salida = [10, 15, 20, 25, 27, 29, 30, 32, 33, 35, 37, 40, 48, 93]$

**Pista 3:**

Consideren las operaciones de apilar y desapilar elementos de las listas.

```python
import heapq

class LSL:
    def __init__(self, valor=0, siguiente=None):
        self.valor = valor
        self.siguiente = siguiente

    def __lt__(self, otro):
        return self.valor < otro.valor

def unir_listas(listas):
    aux = LSL()
    actual = aux

    heap = []

    for l in listas:
        if l:
            heapq.heappush(heap, l)

    while heap:
        node = heapq.heappop(heap)

        actual.siguiente = node
        actual = actual.siguiente

        if node.siguiente:
            heapq.heappush(heap, node.siguiente)

    return aux.siguiente



```

**Explicar el procedimiento que realiza el algoritmo:**

En este algoritmo se busca unir varias listas enlazadas ordenadas, en una sola lista enlazada ordenada utilizando una cola de prioridad (min-heap).

Inicialización:
Se crea un nodo auxiliar "aux" que servirá como el punto de partida para la lista enlazada resultante, "actual" es un puntero que se usa para construir la lista enlazada resultante a partir de aux.

Preparación del Heap:
Inicializa una cola de prioridad "heap" vacía, Itera sobre cada lista en "listas". Si la lista no está vacía, inserta el nodo cabeza de la lista en el heap utilizando heapq.heappush.

Construcción de la Lista Enlazada Resultante:
Mientras el heap no esté vacío: Extrae el nodo con el valor mínimo del heap utilizando "heapq.heappop", Adjunta este nodo a la lista resultante actualizando "actual.siguiente",Mueve el puntero "actual" al nodo recién añadido, Si el nodo extraído tiene un nodo siguiente, inserta este nodo siguiente en el heap.

Resultado:
La función retorna "aux.siguiente", que es la cabeza de la lista enlazada resultante.


```python
# Prueba 1

L1 = LSL(10, LSL(20, LSL(30, LSL(40))))
L2 = LSL(15, LSL(25, LSL(35)))
L3 = LSL(27, LSL(29, LSL(37, LSL(48, LSL(93)))))
L4 = LSL(32, LSL(33))
LSL_unida = unir_listas([L1, L2, L3,L4])

while LSL_unida:
    print(LSL_unida.valor, end=" ")
    LSL_unida = LSL_unida.siguiente
```

```python
# Prueba 2

L1 = None, None, LSL(2, LSL(5, LSL(6)))
L2 = LSL(15), None, LSL(6, LSL(7, LSL(8)))
L3 = None
LSL_unida = unir_listas([L1, L2, L3])

while LSL_unida:
    print(LSL_unida.valor, end=" ")
    LSL_unida = LSL_unida.siguiente
```

## Problema 3 (30 %)

Considerando un árbol binario, diseñe un algoritmo que permita transformalo en una LDL. La conversión debe realizarse de tal forma que los punteros izquierdo y derecho de los nodos del árbol binario actúen como punteros anterior y siguiente en una LDL, y los nodos de la LDL sigan el mismo orden de los nodos de acuerdo con la trayectoria en orden del árbol.

```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izq = None
        self.der = None

class LDL:
    def __init__(self):
        self.cabecera = None
        self.cola = None

    def conectar_nodos(self, nodo1, nodo2):
        nodo1.der = nodo2
        nodo2.izq = nodo1

    def arbol_binario_a_LDL(self, nodo):
        if nodo is None:
            return

        self.arbol_binario_a_LDL(nodo.izq)

        if self.cabecera is None:
            self.cabecera = nodo
        else:
            self.conectar_nodos(self.cola, nodo)

        self.cola = nodo

        self.arbol_binario_a_LDL(nodo.der)

    def mostrar_LDL(self):
        nodo_actual = self.cabecera
        while nodo_actual:
            print(nodo_actual.valor, end=" <--> ")
            nodo_actual = nodo_actual.der
        print("None")
```

**Explicación del algoritmo propuesto:**

*Método conectar_nodos*

Este método se encarga de conectar dos nodos en una lista doblemente ligada:

* `nodo1.der = nodo2`: Establece que el nodo nodo1 apunta al nodo nodo2 como su siguiente nodo (der).
* `nodo2.izq = nodo1`: Establece que el nodo nodo2 apunta al nodo nodo1 como su nodo anterior (izq).

En resumen, conectar_nodos asegura que nodo1 esté antes de nodo2 en la lista doblemente ligada.

---
*Método arbol_binario_a_LDL*


Este método convierte un árbol binario en una lista doblemente ligada, utilizando un recorrido inorden:

`Caso Base`:

Si nodo es None, retorna.

`Recorrido Inorden`:

* Primero visita el subárbol izquierdo (self.arbol_binario_a_LDL(nodo.izq)).

* Luego conecta el nodo actual (nodo) a la lista doblemente ligada:

  * Si self.cabecera es None, establece nodo como la cabecera.
  * Si no es None, conecta el nodo actual (nodo) con el último nodo visitado (self.cola).

* Actualiza self.cola para que apunte al nodo actual (nodo).

* Finalmente, visita el subárbol derecho (self.arbol_binario_a_LDL(nodo.der)).

En resumen, arbol_binario_a_LDL convierte un árbol binario en una lista doblemente enlazada, asegurando que los nodos estén en orden inorden en la lista.


```python
# Prueba 1

raiz = Nodo(1)
raiz.izq = Nodo(2)
raiz.der = Nodo(3)
raiz.izq.izq = Nodo(4)
raiz.izq.der = Nodo(5)
raiz.der.izq = Nodo(6)
raiz.der.der = Nodo(7)

lista = LDL()
lista.arbol_binario_a_LDL(raiz)

print("LDL = ", end="")
lista.mostrar_LDL()
```

```python
# Prueba 2

raiz = Nodo(4)
raiz.izq = Nodo(2)
raiz.der = Nodo(1)
raiz.izq.izq = Nodo(6)
raiz.izq.der = Nodo(7)
raiz.der.izq = Nodo(10)
raiz.der.der = Nodo(11)

lista = LDL()
lista.arbol_binario_a_LDL(raiz)

print("LDL = ", end="")
lista.mostrar_LDL()
```

## Problema 4 (15 %)

**Descripción:**

Considere un árbol binario diseñado para realizar una búsqueda binaria. A partir del árbol, diseñar un algoritmo que permita encontrar el ancestro común más bajo (lowest common ancestor - LCA en inglés) de dos nodos dados por el árbol de búsqueda binaria.

**Pista 1:**

LCA esta definido entre dos nodos, $p$ y $q$, tal que el nodo más bajo en el árbol binario para búsqueda binaria tiene tanto $p$ y $q$ como descendientes (donde se permite que un nodo sea un nodo descendiente a si mismo).

**Pista 3:**

Entrada = [6,2,8,0,4,7,9,null,null,3,5], p = 2, q = 4

Salida: 2

En este caso, el LCA de los nodos 2 y 4 es 2 ya qye un nodo puede ser un descendiente de si mísmo de acuerdo con la definición de LCA.

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def create_binary_tree(values):
    if not values:
        return None
    nodes = [None if value is None else TreeNode(value) for value in values]
    children = nodes[::-1]
    root = children.pop()
    for node in nodes:
        if node:
            if children: node.left = children.pop()
            if children: node.right = children.pop()
    return root

def find_lowest_common_ancestor(root, val1, val2):
    while root:
        if val1 > root.value and val2 > root.value:
            root = root.right
        elif val1 < root.value and val2 < root.value:
            root = root.left
        else:
            return root

def LCA(x, p, q):
    root = create_binary_tree(x)
    lca_node = find_lowest_common_ancestor(root, p, q)
    Minimo_LCA = lca_node.value if lca_node else None
    return Minimo_LCA
```

**Explicación del algoritmo propuesto:**
Es un arbol binario de busqueda, definimos la clase representando un
nodo de árbol binario a partir de una lista de valores y luego determina el ancestro común más cercano para dos valores específicos. En particular, se asume que los valores p y q están en el árbol. Si no están presentes, el resultado será None.












```python
# Prueba 1:

x =  [6,2,8,0,4,7,9,None,None,3,5]
p, q = 2, 8
print("Valor LCA es ", LCA(x, p, q))
```

```python
# Prueba 2:

x =  [6,2,8,0,4,7,9,None,None,3,5]
p, q = 2, 4
print("Valor LCA es ", LCA(x, p, q))
```

## Problema 5 (20 %)

**Descripción:**

Una trayectoria de un árbol binario es una secuencia de nodos donde cada par de nodos adyacentes en la secuencia tiene una arista que los conecta. Un nodo sólo puede aparecer en la secuencia como máximo **una vez**. No es necesario que el camino pase por el nodo raíz.

A partir de lo anterior, la trayectoria es la suma de los valores de los nodos en el trayecto. Considerando el nodo raíz de un árbol binario, proponer un algoritmo que retorne la suma máxima de cualquier trayectoria no vacío.

**Pista 1:**

Entrada = [1,2,3]

Salida: 6

Análisis: La trayectoria es 2 -> 1 -> 3 cuya suma máxima es 2 + 1 + 3 = 6.

**Pista 2:**

Entrada = [-10,9,20,None,None,15,7]

Salida: 42

Análisis: La trayectoria es 15 -> 20 -> 7 cuya suma máxima es 15 + 20 + 7 = 42.

**Pista 3:**

El primer valor del arreglo es el nodo raíz.

```python
class Nodo:
    def __init__(self, valor, izquierdo=None, derecho=None):
        self.valor = valor
        self.izquierdo = izquierdo
        self.derecho = derecho

def construir_arbol(valores, indice=0):
    if indice >= len(valores) or valores[indice] is None:
        return None

    valor = valores[indice]
    nodo = Nodo(valor)

    nodo.izquierdo = construir_arbol(valores, 2 * indice + 1)
    nodo.derecho = construir_arbol(valores, 2 * indice + 2)

    return nodo


def suma_maxima(nodo):
    def funcion_auxiliar(nodo):
        nonlocal max_suma
        if nodo is None:
            return 0

        izquierda = max(funcion_auxiliar(nodo.izquierdo), 0)  # ignoramos sumas negativas
        derecha = max(funcion_auxiliar(nodo.derecho), 0)  # ignoramos sumas negativas

        # Suma máxima en el camino que pasa por el nodo actual
        suma_actual = nodo.valor + izquierda + derecha

        # Actualizamos la suma máxima global si es necesario
        max_suma = max(max_suma, suma_actual)

        # Retornamos la suma máxima de un camino que termina en el nodo actual
        return nodo.valor + max(izquierda, derecha)

    max_suma = 0
    funcion_auxiliar(nodo)
    return max_suma
```

**Explicar el procedimiento que realiza el algoritmo:**

**Función construir_arbol:**
Esta función construye un árbol binario a partir de una lista de valores.

La función toma dos argumentos:

* `valores: La lista que contiene los valores de los nodos del árbol.`
* `indice: El índice actual en la lista de valores.`

Si el índice es mayor o igual que la longitud de la lista o el valor en el índice es None, se devuelve None indicando que no hay más nodos para crear.

Se crea un nuevo nodo con el valor en el índice actual.

Se llama recursivamente a la función `construir_arbol` para crear los subárboles izquierdo y derecho del nodo actual, pasando como argumentos el índice para el hijo izquierdo (2 * indice + 1) y el índice para el hijo derecho (2 * indice + 2).

Se establecen los punteros izquierdo y derecho del nodo actual a los nodos creados recursivamente.

Se devuelve el nodo actual.

---
**Función suma_maxima:**
Esta función calcula la suma máxima de un camino en un árbol binario. La función utiliza una función auxiliar recursiva para recorrer el árbol y calcular las sumas máximas para cada nodo. La función toma un único argumento:

* `nodo: El nodo raíz del árbol binario.`

Se define una función auxiliar recursiva funcion_auxiliar que toma un nodo como argumento.

Se declara una variable `max_suma` para almacenar la suma máxima global encontrada hasta el momento. Se inicializa en 0.

Se llama a la función auxiliar `funcion_auxiliar` con el nodo raíz.

Se devuelve el valor de `max_suma`.

La función auxiliar `funcion_auxiliar` funciona de la siguiente manera:

Si el nodo es None, se devuelve 0.

Se calculan las sumas máximas para los subárboles izquierdo y derecho del nodo actual utilizando llamadas recursivas a `funcion_auxiliar`.

Se ignora cualquier suma negativa proveniente de los subárboles.

Se calcula la suma máxima para el camino que pasa por el nodo actual: suma_actual = nodo.valor + izquierda + derecha.

Se actualiza la variable `max_suma` si la suma actual es mayor que la suma máxima global encontrada hasta el momento.

Se devuelve la suma máxima de un camino que termina en el nodo actual: nodo.valor + max(izquierda, derecha).

En resumen, el código define las clases y funciones necesarias para construir un árbol binario a partir de una lista de valores y calcular la suma máxima de un camino en el árbol. La función `construir_arbol` utiliza un recorrido previo para crear la estructura del árbol, mientras que la función `max_suma` utiliza una función auxiliar recursiva para recorrer el árbol y calcular las sumas máximas para cada nodo.


```python
# Prueba 1
x = [1,2,3]
arbol = construir_arbol(x)
a = suma_maxima(arbol)
print("La Suma Máxima es: ", a)
```
